```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For better visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
# Step 2: Create a sample dataset
data = {
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.randint(20000, 120000, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing'], 100)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Step 3: Basic information about data
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Step 4: Univariate Analysis (Numerical - Histogram)
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 5: Univariate Analysis (Numerical - Boxplot)
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Salary'], color='orange')
plt.title('Salary Distribution (Boxplot)')
plt.ylabel('Salary')
plt.show()
```

```python
# Step 6: Univariate Analysis (Categorical - Countplot)
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Step 7: Univariate Analysis (Categorical - Pie Chart)
dept_counts = df['Department'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Department Distribution')
plt.show()
```
