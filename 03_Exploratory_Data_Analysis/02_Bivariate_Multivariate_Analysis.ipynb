```python

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
# Step 2: Create a Sample Dataset
np.random.seed(42)
data = {
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.randint(20000, 120000, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing'], 100),
    'Purchase_Amount': np.random.randint(1000, 20000, 100)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Step 3: Quick Overview
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nSummary:\n", df.describe(include='all'))
```

```python
# Step 4: Scatter Plot — Relationship between Age & Salary
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Salary', hue='Gender', data=df, palette='Set2')
plt.title('Age vs Salary (Colored by Gender)')
plt.show()
```

```python
# Step 5: Bar Plot — Average Salary by Department
plt.figure(figsize=(7,5))
sns.barplot(x='Department', y='Salary', data=df, palette='pastel', estimator='mean')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```

```python
# Step 6: Violin Plot — Salary Distribution by Gender
plt.figure(figsize=(7,5))
sns.violinplot(x='Gender', y='Salary', data=df, palette='muted')
plt.title('Salary Distribution by Gender')
plt.show()
```

```python
# Step 7: Correlation Matrix (Numerical Columns)
corr = df.corr(numeric_only=True)
print(corr)

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Step 8: Pairplot — Relationships among multiple numerical variables
sns.pairplot(df, hue='Gender', palette='husl')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

```python
# Step 9: Groupby Comparison — Average Purchase Amount by Gender & Department
group_data = df.groupby(['Gender', 'Department'])['Purchase_Amount'].mean().unstack()
print(group_data)

group_data.plot(kind='bar', figsize=(8,5), colormap='Pastel1')
plt.title('Average Purchase Amount by Gender and Department')
plt.ylabel('Average Purchase Amount')
plt.show()
```
