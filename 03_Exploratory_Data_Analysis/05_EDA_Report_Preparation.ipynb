```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
# Step 2: Create or Load Dataset
np.random.seed(1)
data = {
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.randint(25000, 120000, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing'], 100),
    'Purchase_Amount': np.random.randint(1000, 25000, 100)
}

df = pd.DataFrame(data)
df.head()
```

```python
# Step 3: Dataset Overview
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nℹDataset Info:")
print(df.info())
```

```python
# Step 4: Missing Values Check
print("Missing Values:\n", df.isnull().sum())
```

```python
# Step 5: Summary Statistics
print("Summary Statistics:\n")
print(df.describe(include='all'))
```

```python
# Step 6: Pairplot for Relationship Overview
sns.pairplot(df, hue='Gender', palette='husl')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

```python
# Step 7: Correlation Heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Example Summary
print("""
 EDA Summary:
-----------------------------
• Dataset has 100 rows and 5 columns.
• No missing values detected.
• Numerical features: Age, Salary, Purchase_Amount.
• Categorical features: Gender, Department.
• Age & Salary roughly normally distributed.
• Salary correlates moderately with Age (~0.5).
• Purchase_Amount shows high variance across departments.
• Gender balance is nearly equal.
-----------------------------
""")
```
