```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

# 1. Dataset: Titanic (Survival Prediction)

```python
# Step 2: Load Titanic Dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# Step 3: Basic Info & Missing Values
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

EDA Q1 â€” Are there missing or skewed data?

```python
# Visualize Missing Data
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()

# Summary statistics
df.describe(include='all')
```

EDA Q2 â€” Which feature affects survival most?

```python
# Survival Rate by Gender
sns.barplot(x='sex', y='survived', data=df, palette='pastel')
plt.title("Survival Rate by Gender")
plt.show()

# Survival Rate by Class
sns.barplot(x='class', y='survived', data=df, palette='muted')
plt.title("Survival Rate by Passenger Class")
plt.show()

# Survival Rate by Embarkation
sns.barplot(x='embarked', y='survived', data=df, palette='cool')
plt.title("Survival Rate by Embarkation Port")
plt.show()
```

EDA Q3 â€” What patterns exist between variables?

```python
# Pairplot for numerical relationships
sns.pairplot(df[['age', 'fare', 'survived']], hue='survived', palette='husl')
plt.suptitle("Pairwise Relationships (Titanic)", y=1.02)
plt.show()

# Correlation heatmap
corr = df[['age', 'fare', 'survived']].corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Heatmap (Titanic)")
plt.show()
```

# 2. Dataset: Iris (Flower Classification)

```python
# Load Iris Dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Basic Overview
print(iris.info())
print(iris.describe())
```

```python
# Distribution of features
for col in iris.columns[:-1]:
    plt.figure(figsize=(6,4))
    sns.histplot(iris[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Pairwise relationships
sns.pairplot(iris, hue='species', palette='Set2')
plt.suptitle("Iris Feature Relationships", y=1.02)
plt.show()

# Correlation heatmap
sns.heatmap(iris.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Iris)")
plt.show()
```

```python
# Which feature affects species most?
sns.boxplot(x='species', y='petal_length', data=iris, palette='pastel')
plt.title("Petal Length vs Species")
plt.show()
```
