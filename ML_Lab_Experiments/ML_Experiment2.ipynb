```python
import pandas as pd

# Load dataset (CSV file)
df = pd.read_csv("/content/Student_Dropout_Dataset.csv")

# See first 5 rows
print(df.head())

# Check data types and missing values
print(df.info())

# Basic statistics
print(df.describe())
```

```python
import pandas as pd
import numpy as np

def generate_data_dictionary(df):
    dd = pd.DataFrame({
        'feature': df.columns,
        'dtype': [str(df[c].dtype) for c in df.columns],
        'n_unique': [df[c].nunique(dropna=False) for c in df.columns],
        'n_missing': [int(df[c].isna().sum()) for c in df.columns],
        'pct_missing': [round(df[c].isna().mean()*100,2) for c in df.columns],
        'sample_values': [list(df[c].dropna().unique()[:5]) for c in df.columns],
        'description': ['' for _ in df.columns],   # you fill this in
        'relevance': ['' for _ in df.columns],     # you fill this in
        'notes': ['' for _ in df.columns],
    })
    return dd

dd = generate_data_dictionary(df)
display(dd)
dd.to_csv("/content/Student_Dropout_Dataset.csv", index=False)
```
